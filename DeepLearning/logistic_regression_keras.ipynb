## kerasによる実装
Kerasでは$x$, $y$などを考える必要がなくなる。

````python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
````

```python
Dense(input_dim=2, units=1)
```
を定義することは、以下の式に相当する層を作ったことになる
$$
w_1x_1 + w_2x_2 + b
$$
ニューロンの出力を表現するためには、ここに活性化関数が必要になるので、
```python
Activation('sigmoid')
```
によって、
$$
y = \sigma(w_1x_1 + w_2x_2 + b)
$$
を表現する層を作る

````python
model = Sequential([
    Dense(input_dim=2, units=1),
    Activation('sigmoid')
])
````

以下のような記述も可能。model.add()でどんどん層を追加していく方法
```python
model = Sequential()
model.add(Dense(input_dim=2, units=1))
model.add(Activation('sigmoid'))
```

````python
# 確率的勾配降下法
# lrは学習率(learning rate)
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
````

````python
# データの準備
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [1]])
````

モデルの学習はKerasでは以下のように
```python
model.fit()
```
を実行するだけ
epochsはエポック数、batch_sizeはミニバッチの大きさをそれぞれ指定している

````python
model.fit(X, Y, epochs=200, batch_size=1)
````

````python
classes = model.predict_classes(X, batch_size=1)
prob = model.predict_proba(X, batch_size=1)

print('classified:')
print(Y == classes)
print()
print('output probability')
print(prob)
````
