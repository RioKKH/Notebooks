```r
library("tidyverse")
```

```r
n <- seq(1, 5, length.out = 100)
logn <- log(n)
rootn <- sqrt(n)
nlogn <- n * log(n) 
nsquare <- n^2
power2n <- 2^n
factorialn <- gamma(n + 1)
```

```r
df <- data.frame(n = n, logn = logn, rootn = rootn, nlogn = nlogn, nsquare = nsquare,
                 power2n = power2n, factorialn = factorialn) 
```

```r
dfm <- df %>% 
    #tidyr::gather(key = kind, value = value, c(logn, rootn, nlogn, nsquare, power2n, fctorialn))
    tidyr::gather(key = kind, value = value, logn:factorialn)
```

```r
ggplot(data = dfm) +
    geom_point(mapping = aes(x = n, y = value, color=kind)) +
    geom_line(mapping = aes(x = n, y = value, color=kind))
```
